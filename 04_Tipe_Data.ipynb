# **Tipe Data Primitif**

Tipe data primitif merupakan jenis paling dasar dalam pemrograman. Tipe data ini menyimpan single value. Berikut adalah berbagai tipe data primitif.

### Numbers

*   **Integer**: Bilangan bulat positif atau negatif dan tidak memiliki angka desimal.
*   **Float**: Bilangan riil yang dapat mewakili bilangan bulat atau bilangan desimal.
*   **Complex**: Bilangan kompleks.

````python
x = 6
print(type(x))

x = 6.0
print(type(x))

x = 1+2j
print(type(x))
````

````python
var = 10
print(var)
print(id(var))

var = 11
print(var)
print(id(var))
````

### Boolean



*   **True**: Bernilai benar
*   **False**: Bernilai salah



````python
x = True
print(type(x))

x = False
print(type(x))
````

### String

String merupakan karakter yang berurutan. Ketika Anda membuat variabel bernilai string tentu diawali dengan single quote (‘’) atau double quote (“”).

````python
x = 'Dicoding'
print(type(x))
````

Anda dapat menggunakan tiga single quote atau double quote untuk menyimpan string yang lebih dari satu baris (multi-line).

````python
multi_line = """Halo!
Kapan terakhir kali kita bertemu?
Kita bertemu hari Jum’at yang lalu."""

print(multi_line)
````

Anda dapat mengakses setiap karakter dari string (substring) dengan menggunakan metode indexing. Perlu diingat bahwa indeks selalu dimulai dari 0.

````python
x = 'Dicoding'
print(x[0])
````

Namun, Anda tidak dapat mengubah substring di dalamnya. Ini dikarenakan string pada Python bersifat immutable.

````python
x = 'Dicoding'
x[0] = 'F'
````

Anda dapat mengakses beberapa substring dengan menggunakan metode indexing dan slicing.

````python
x = 'Dicoding'
print(x[2:])
````

Anda dapat menampilkan teks/string berdasarkan input dari pengguna dengan berbagai cara.

````python
# Formatted String

name = "Nugie Saputra"
print(f"Hello, nama saya {name}")
````

````python
# %-formating
name = "Perseus Evans"
print("Nama saya %s" % (name))
````

````python
name = "Arsha"
print("Nama saya {}".format(name))
````

# **Tipe Data Collection**

## List

List merupakan jenis kumpulan data terurut (ordered sequence) dan salah satu tipe data yang sering digunakan pada Python. List Python tidak mengharuskan memiliki tipe data yang sama di dalamnya, sedangkan array sebaliknya.

```
variabel_list = [var1, var2, var3]
```

````python
x = [1, 2.2, 'Dicoding']
print(type(x))
````

````python
x = [1, 'Dicoding', True, 1.0]

print(x[2])
````

List Python bersifat mutable yang artinya nilai di dalamnya dapat diubah.

````python
x = [1, 2.2, 'Dicoding']
x[0] = 'Indonesia'
print(x)
````

Konsep indexing merujuk kepada pengambilan data dalam Python berdasarkan indeksnya. Beberapa cara untuk melakukan indexing sebagai berikut.

````python
x = ["laptop", "monitor", "mouse", "mousepad", "keyboard", "webcam", "microphone"]

print(x[0])
print(x[2])
print(x[-1])
print(x[-3])
````

Adapun konsep slicing merujuk pada pengambilan data berdasarkan indeks dari rentang tertentu. Slicing pada Python mengikuti pola sebagai berikut.

```
sequence[start:stop:step]
```

````python
x = ["laptop", "monitor", "mouse", "mousepad", "keyboard", "webcam", "microphone"]

print(x[0:5:2])
print(x[1:])
print(x[:3])
````

## Tuple

Tuple adalah jenis dari list yang tidak dapat diubah elemennya. Umumnya, tuple digunakan untuk data yang bersifat sekali deklarasi dan dapat dieksekusi lebih cepat.

````python
x = (1, "Dicoding", 1+3j)
print(type(x))
````

````python
x = (5, 'program', 1+3j)
print(x[1])
print(x[0:3])
````

Tuple bersifat immutable yang artinya tidak dapat diubah.

````python
x = (5, 'program', 1+3j)
x[1] = 'Dicoding'
````

## Set

Set adalah kumpulan item bersifat unik, tanpa urutan (unordered collection). Tidak sama seperti list, dalam set kita tidak bisa melakukan indeksing karena set tidak memiliki indeks. Hal ini merujuk pada definisi nya yang menyatakan bahwa set merupakan kumpulan item tanpa urutan.

````python
x = {1,2,7,2,3,13,3}
print(x[0])
````

Pada kode di atas, program mengembalikan output 'set' object is not subscriptable karena setiap nilai dalam set tidak memiliki indeks sehingga tidak bisa dilakukan indexing.

````python
x = {1, 2, 7, 2, 3, 13, 3}
print(x)
print(type(x))
````

Terakhir, set adalah himpunan dalam matematika. Ini maknanya Anda dapat melakukan operasi union (gabungan) dan intersection (irisan) pada set. Python menyediakan method “.union()” dan “.intersection()” untuk tipe data set.

**Method** merupakan tindakan atau operasi yang dapat dilakukan oleh suatu objek. Saat ini, tidak apa-apa jika kamu belum memahami sepenuhnya. Anda akan mempelajari lebih detail mengenai method pada modul Object-Oriented Programming (OOP).

**intersection** atau irisan yang merupakan operasi dalam matematika untuk menemukan nilai atau elemen-elemen yang sama dalam dua atau lebih himpunan.

````python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

union = set1.union(set2)
print("Union:", union)

intersection = set1.intersection(set2)
print("Intersection:", intersection)
````

### Dictionary

Dictionary pada Python merupakan kumpulan pasangan key-value yang bersifat tidak berurutan. Dictionary dapat digunakan untuk menyimpan data kecil hingga besar.

```
variabel = {'key': 'value'}
```

````python
x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }

print(type(x))
````

Dalam mengambil setiap nilai/elemen pada dictionary, Anda harus mengetahui key (kunci) untuk mengakses setiap value-nya (nilai). Hal ini berbeda dengan tipe data sebelumnya yang cukup mengharuskan Anda untuk menyebutkan indeksnya saja.

````python
x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }

print(x[0])
````

Kode di atas menghasilkan error karena Anda mencoba mengakses data pada dictionary dengan menggunakan metode indexing.

````python
x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }

print(x ['name'])
````

Dalam kode di atas, Anda mengambil data pada dictionary dengan memanggil key yang ada.

````python
# Menambah Data pada Dictionary

x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }
x ['Job'] = "Web Developer"

print(x)
````

````python
# Menghapus Data pada Dictionary

x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }
del x['isMarried']

print(x)
````

````python
# Mengubah Data pada Dictionary

x = { 'name': 'Perseus Evans', 'age': 20, 'isMarried': False }
x ['name'] = "Dicoding"

print(x)
````

# **Konversi antara Tipe Data**

````python
# Konversi Integer ke Float
print(float(5))
````

````python
# Konversi Float ke Integer
print(int(5.6))
print(int(-5.5))
````

````python
# Konversi dari-dan-ke String
print(int("25"))
print(str(25))
print(float("25"))
print(str(25.6))
````

````python
# Konversi Kumpulan Data
print(set([1,2,3]))
print(tuple({5,6,7}))
print(list('hello'))
````

````python
# Konversi ke Dictionary
print(dict([[1,2],[3,4]]))
````

````python
print(dict([(3,26),(4,44)]))
````
